```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.multivariate_normal(mean=[-10, 10], cov=[[10, 1], [1, 10]], size=1000))
df = df.rename(columns={0:'Feature1', 1:'Feature2'})

df['Response'] = np.random.rand(1) * df.Feature1 + np.random.rand(1) * df.Feature2
df = df[['Response', 'Feature1', 'Feature2']]
```

```python
# Shape of the Dataframe.
df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Descriptive statistics and a visual of the three columns
sns.histplot(data=df, x='Response');
plt.show()
sns.histplot(data=df, x='Feature1');
plt.show()
sns.histplot(data=df, x='Feature2');
```

```python
# Scatterplot depicting the linear correlated nature of the three columns.
sns.scatterplot(data=df, x='Feature1', y='Response');
plt.show()
sns.scatterplot(data=df, x='Feature2', y='Response');
```

```python
# Correlation coefficients of the three pairs (Response with Feature1 etc.).
df.corr()
```
